```python
import numpy as np
l = [2,21,88]
arr = np.array(l)

def is_prime(data):
    middle_num = int(data **0.5)
    for i in range(2,middle_num+1):
        if data % i == 0:
            return False
    return True

bool_index = [is_prime(data) for data in arr]
# print(bool_index)
arr[bool_index]
```

```python
# Marvin 

import numpy as np

def is_prime(value):
    end = np.floor(np.sqrt(value)+1)
    for i in range(2, int(end)):
        if value % i ==0:
            return False
    return True

random_arr = np.random.randint(0,299,20)
print(random_arr)
is_prime_result = [is_prime(value) for value in random_arr] 
random_arr[is_prime_result]
```

```python
# 课件
import numpy as np

def is_prime(x):
    div_cnt = 0
    for i in range(1, x+1):
        if x % i == 0:
            div_cnt += 1
        if div_cnt > 2:
            break
    return div_cnt == 2

is_prime_ufunc = np.vectorize(is_prime)
```
